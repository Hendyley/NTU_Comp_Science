```python
# Basic Libraries
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

# P1 DataPREP

```python

#a and b
train = pd.read_csv("train.csv")

print("Data type : ", type(train))
print("Data dims : ", train.shape)
print(train.dtypes)
```

```python
#c

trainint = train.select_dtypes("int")
#print(trainint)
trainint.info()
```

```python
#d n e
trainint = trainint.drop(columns=['MSSubClass', 'OverallQual', 'OverallCond'])
trainint.info()
```

```python
trainint
```

# P2  Statistical Summary

```python
#a
salesprice = pd.DataFrame(trainint[["SalePrice"]])
salesprice.describe().round(2)
```

```python
#b
# Draw the distributions of all variables
f, axes = plt.subplots(1, 3, figsize=(24, 10))
#plt.xticks(rotation=90)


sb.boxplot(data = salesprice, orient = "h", ax = axes[0])
sb.histplot(data = salesprice, ax = axes[1])
sb.violinplot(data = salesprice, orient = "h", ax = axes[2])
```

```python
#c n d
LotArea = pd.DataFrame(trainint[["LotArea"]])
LotArea.describe().round(2)

# Draw the distributions of all variables
f, axes = plt.subplots(1, 3, figsize=(25, 6))
#plt.xticks(rotation=90)


sb.boxplot(data = LotArea, orient = "h", ax = axes[0])
sb.histplot(data = LotArea, ax = axes[1])
sb.violinplot(data = LotArea, orient = "h", ax = axes[2])
```

```python
#e
# Create a joint dataframe by concatenating the two variables
jointDF = pd.concat([LotArea,salesprice], axis = 1).reindex(LotArea.index)
jointDF

```

```python
# Draw jointplot of the two variables in the joined dataframe
sb.jointplot(data = jointDF, x = "LotArea", y = "SalePrice", height = 12)
```

```python
# Calculate the correlation between the two columns/variables
jointDF.corr()
```
