```python
# New Class to take in different depth and width

class Part_A3_MLP(nn.Module):

    def __init__(self, no_features, hidden_units, no_labels):
        super().__init__()

        depth = len(hidden_units)
        
        layers = []
        
        # Adding hidden layers
#         print("1 hidden \n",no_features, hidden_units[0])
        layers.append(nn.Linear(no_features, hidden_units[0]))
        layers.append(nn.ReLU())
        layers.append(nn.Dropout(0.2))
        
        if depth > 1: 
            for i in range(depth - 1):
#                 print("1 hidden \n",hidden_units[i], hidden_units[i+1])
                layers.append(nn.Linear(hidden_units[i], hidden_units[i+1]))
                layers.append(nn.ReLU())
                layers.append(nn.Dropout(0.2))

        # Adding output layer
        layers.append(nn.Linear(hidden_units[-1], no_labels))
        layers.append(nn.Sigmoid())
        
        self.mlp_stack = nn.Sequential(*layers)

    def forward(self, x):
        return self.mlp_stack(x)
```

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    model.train()
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    train_loss, correct = 0, 0
    for batch, (X, y) in enumerate(dataloader):
        # Compute prediction and loss
        pred = model(X)
        loss = loss_fn(pred, y)
        
        train_loss += loss.item()
        correct += ((pred > 0.5).type(torch.float) == y).type(torch.float).sum().item()


        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        
        # Update Weight
        optimizer.step()
    
        if batch % 100 == 0:
                loss, current = loss.item(), (batch + 1) * len(X)
                print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
    
    train_loss /= num_batches
    correct /= size
    print(f"Train Error: \n Accuracy: {(correct*100):>0.1f}%, Avg loss: {train_loss:>8f} \n")
    
    return train_loss, correct
    

def test_loop(dataloader, model, loss_fn):
    model.eval()
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += ((pred > 0.5).type(torch.float) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
    
    return test_loss, correct
```

3.Perform hyperparameter tuning for the different neurons with 5-fold cross validation.

```python
def train(model, X_train_scaled, y_train2, X_val_scaled, y_val2, batch_size):

    # YOUR CODE HERE
    
    train_data = CustomDataset(X_train_scaled, y_train2)
    val_data = CustomDataset(X_val_scaled, y_val2)

    train_dataloader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    val_dataloader = DataLoader(val_data, batch_size=batch_size, shuffle=True)
    num_batches = len(train_dataloader)
    
    loss_fn = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
    Early_stopper = EarlyStopper(patience=3, min_delta=0)


    train_losses, train_accuracies = [], [] 
    test_losses, test_accuracies = [], []
    accuracies, total_times, time_last_epoch = [], [], []
    t = 0

    for epoch in range(epochs):

        print(f"Epoch {epoch+1}\n-------------------------------")

        start = time.time()

        train_loss, train_accuracy = train_loop(train_dataloader, model, loss_fn, optimizer)
        test_loss, test_accuracy = test_loop(val_dataloader, model, loss_fn)

        end =  time.time()
        t +=  end - start

        train_losses.append(train_loss), train_accuracies.append(train_accuracy)
        test_losses.append(test_loss), test_accuracies.append(test_accuracy)

        # Early Stopper
        if Early_stopper.early_stop(test_loss): 
            print("Done!")
            break

        # Accuracy for the current Fold at last Epoch
        accuracies.append(test_accuracies[-1])

        # Total Time taken 
        total_times.append(t)#*1000 / (epochs) )
        
        # Time taken on last epoch
        time_last_epoch.append(end - start)

        print(f"Batch size: {batch_size}, Time taken: {total_times[-1]:>3f}s , Accuracy: {accuracies[-1]:>2f} , Time taken last epoch: {time_last_epoch[-1]:>3f}s ")       

    return train_accuracies, train_losses, test_accuracies, test_losses, time_last_epoch
```

```python
def find_optimal_hyperparameter(X_train, y_train, parameters, mode, batch_size):

    # YOUR CODE HERE
    
    # 5-fold cross validation
    kf = KFold(n_splits=5, shuffle=True, random_state=0)
    
    cross_validation_accuracies = []
    cross_validation_times = []
    
 
    for parameter in parameters:
        print(f"Parameter {parameter}")
        accuracies = []
        times = []

        for train_index, val_index in kf.split(X_train):
            X_train2, X_val2 = X_train[train_index], X_train[val_index]
            y_train2, y_val2 = y_train[train_index], y_train[val_index]

            model = Part_A3_MLP(X_train_scaled.shape[1],parameter,1)

            train_accuracy_, train_loss_, test_accuracy_, test_loss_, time = train(model, X_train2, y_train2, X_val2, y_val2, batch_size)
            
            # Save the accuracy for each fold at the last epoch
            accuracies.append(test_accuracy_[-1])
            
            # Save the time taken to train
            times.append(time[-1])
            
        # Mean Accuracy for the Number of Neurons (Mean of Accuracy at Last Epoch for each Fold)
        cross_validation_accuracies.append(np.mean(accuracies))

        # Mean Time taken for the Number of Neurons (Mean of Time at Last Epoch for each Fold)
        cross_validation_times.append(np.mean(times))

    return cross_validation_accuracies, cross_validation_times

'''
optimal_bs = 0. Fill your optimal batch size in the following code.
'''
# YOUR CODE HERE
optimal_bs = 512 # From Part_A2
combo = [[64], [128], [256], [64, 64], [128, 128], [256, 256], [64, 128], [128, 64], [128, 256], [64, 256], [256, 128], [256, 64], [64, 64, 64], [128, 128, 128], [256, 256, 256], [64, 128, 256], [64, 256, 256], [128, 64, 64], [128, 128, 64], [256, 128, 64], [256, 256, 128]]
num_neurons = combo
cross_validation_accuracies, cross_validation_times = find_optimal_hyperparameter(X_train, y_train, num_neurons, 'num_neurons', optimal_bs)
```

## One Better approach is : Randomize search for optimal combination

One approach to efficiently search through combinations of widths and depths is to use randomized search instead of an exhaustive grid search. 

Randomized search randomly samples combinations from the search space, allowing for a more efficient exploration of the hyperparameter space while still providing good performance.

Randomized search is more efficient than searching through all possible combinations because it randomly samples parameter settings from the specified distributions, rather than exhaustively exploring every single combination. This approach can significantly reduce computational time, especially when the search space is large.

While there is a possibility of missing out on the most efficient combination with randomized search, the trade-off is acceptable because randomized search tends to find good parameter settings quickly. It explores a diverse range of combinations, including some that may not have been considered in a grid search, thereby increasing the likelihood of finding good hyperparameters.

Additionally, randomized search allows you to specify the <b> number of iterations (n_iter)</b>, which determines the number of parameter settings to sample. By adjusting this parameter, you can control the balance between search efficiency and the thoroughness of exploration. You can start with a relatively small n_iter value and increase it if necessary to further explore the search space. Overall, randomized search offers a practical compromise between efficiency and effectiveness in hyperparameter optimization.

```python
# YOUR CODE HERE

import random

def Randomized_search_optimal_hyperparameter(X_train, y_train, parameters, mode, batch_size):

    kf = KFold(n_splits=5, shuffle=True, random_state=0)
    
    cross_validation_accuracies = []
    cross_validation_times = []
    
    for parameter in parameters:
        accuracies = []
        times = []

        for train_index, val_index in kf.split(X_train):
            X_train2, X_val2 = X_train[train_index], X_train[val_index]
            y_train2, y_val2 = y_train[train_index], y_train[val_index]

            model = Part_A3_MLP(X_train_scaled.shape[1], parameter, 1)

            train_accuracy_, train_loss_, test_accuracy_, test_loss_, time = train(model, X_train2, y_train2, X_val2, y_val2, batch_size)
            
            accuracies.append(test_accuracy_[-1])
            times.append(time[-1])
            
        cross_validation_accuracies.append(np.mean(accuracies))
        cross_validation_times.append(np.mean(times))

    return cross_validation_accuracies, cross_validation_times

def random_search(parameters, num_trials):
    random_combinations = random.sample(parameters, num_trials)
    return random_combinations

optimal_bs = 512 # From Part_A2
combo = [[64], [128], [256], [64, 64], [128, 128], [256, 256], [64, 128], [128, 64], [128, 256], [64, 256], [256, 128], [256, 64], [64, 64, 64], [128, 128, 128], [256, 256, 256], [64, 128, 256], [64, 256, 256], [128, 64, 64], [128, 128, 64], [256, 128, 64], [256, 256, 128]]
n_iter = 10 # limit number of trial
num_neurons = random_search(combo, n_iter)

Randomized_cross_validation_accuracies, Randomized_cross_validation_times = Randomized_search_optimal_hyperparameter(X_train, y_train, num_neurons, 'num_neurons', optimal_bs)
```
