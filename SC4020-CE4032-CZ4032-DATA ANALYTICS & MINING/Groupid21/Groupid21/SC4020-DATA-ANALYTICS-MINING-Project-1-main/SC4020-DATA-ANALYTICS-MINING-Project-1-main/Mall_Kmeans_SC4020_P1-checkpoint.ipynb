```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
import seaborn as sns
```

```python
mall_df = pd.read_csv("Mall_Customers.csv")
mall_df.head()
```

```python
mall_df.describe()
```

```python
# Encoding the Gender column
mall_df['Gender'] = mall_df['Gender'].map({'Male': 0, 'Female': 1})

mall_df.head()
```

```python
# Select the financial-related metrics features for clustering
money_df = mall_df[['Annual Income (k$)', 'Spending Score (1-100)']]

money_df.head()
```

```python
# Select the dmeographics-related metrics features for clustering
dem_df = mall_df[['Gender', 'Age']]

dem_df.head()
```

```python
# Select the 'hypothesized' major metrics features for clustering
maj_df = mall_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

maj_df.head()
```

```python
# Select all the metrics features for clustering
all_df = mall_df[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

all_df.head()
```

```python
# Correlation matrix for numerical features
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_all = all_df.corr()

# Plot the heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(corr_all, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot for EDA
sns.pairplot(mall_df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Gender']])
plt.show()
```

```python
# Using the elbow method to find the optimal number of clusters
wcss = []  # Within-cluster sum of squares

# Try different values of k (number of clusters)
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(money_df)
    wcss.append(kmeans.inertia_)

# Plot the results to visualize the Elbow
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Within-cluster Sum of Squares)')
plt.show()
```

```python
# Apply K-means to the dataset with k=5
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(money_df)

# Visualize the clusters
plt.figure(figsize=(10, 6))

# Plot each cluster
plt.scatter(money_df.iloc[y_kmeans == 0, 0], money_df.iloc[y_kmeans == 0, 1], s=100, c='red', label='Cluster 1')
plt.scatter(money_df.iloc[y_kmeans == 1, 0], money_df.iloc[y_kmeans == 1, 1], s=100, c='blue', label='Cluster 2')
plt.scatter(money_df.iloc[y_kmeans == 2, 0], money_df.iloc[y_kmeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(money_df.iloc[y_kmeans == 3, 0], money_df.iloc[y_kmeans == 3, 1], s=100, c='cyan', label='Cluster 4')
plt.scatter(money_df.iloc[y_kmeans == 4, 0], money_df.iloc[y_kmeans == 4, 1], s=100, c='magenta', label='Cluster 5')

# Plot the centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

# Labels and title
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments based on Annual Income and Spending Score')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score
sil_score = silhouette_score(money_df, y_kmeans)
print(f'Silhouette Score: {sil_score}')
```

```python
from sklearn.metrics import davies_bouldin_score, calinski_harabasz_score
```

```python
# Calculate Davies-Bouldin score
davies_bouldin = davies_bouldin_score(money_df, y_kmeans)
print(f"Davies-Bouldin Score: {davies_bouldin}")

# Calculate Calinski-Harabasz score
calinski_harabasz = calinski_harabasz_score(money_df, y_kmeans)
print(f"Calinski-Harabasz Score: {calinski_harabasz}")
```

```python
# Add the cluster labels to money_df
money_df['Cluster'] = y_kmeans

# Create a new DataFrame combining Age and Cluster information
age_cluster_df = mall_df[['Age']].copy()
age_cluster_df['Cluster'] = y_kmeans

# Display the first few rows to confirm the merge
age_cluster_df.head()
```

```python
# Calculate descriptive statistics for Age by Cluster
age_stats = age_cluster_df.groupby('Cluster').agg(['mean', 'median', 'std', 'min', 'max'])
print(age_stats)
```

```python
# Boxplot of Age by Cluster
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Age', data=age_cluster_df)
plt.title('Age Distribution for Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Age')
plt.show()

# Histogram of Age by Cluster
plt.figure(figsize=(10, 6))
for cluster in age_cluster_df['Cluster'].unique():
    sns.histplot(age_cluster_df[age_cluster_df['Cluster'] == cluster]['Age'], label=f'Cluster {cluster}', kde=True)
    
plt.title('Age Distribution for Each Cluster')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Create age bands using pandas `cut` function
age_cluster_df['Age Band'] = pd.cut(age_cluster_df['Age'], bins=[10, 19, 29, 39, 49, 59, 69, 79], labels=['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79'])

# Count the number of customers in each age band for each cluster
age_band_counts = age_cluster_df.groupby(['Cluster', 'Age Band']).size().unstack().fillna(0)
print(age_band_counts)

# Visualize the age band distribution for each cluster
age_band_counts.plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title('Age Band Distribution Across Clusters')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.legend(title='Age Bands')
plt.show()
```
